```python
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from google.colab import files
```

```python
#establish connection to google drive - needs to be run from personal account
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/DCC Project/' #change this to match your google drive
```

```python
# import kaggle movie data
movies = pd.read_csv(file_path + 'cleaned_movies.csv')
songs = pd.read_csv(file_path + 'cleaned_songs.csv')
```

```python
songs.head()
```

```python
movies.head()
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
lists = movies["Plot"].tolist()
lists.append(songs["lyrics"])
embedings = model.encode(lists)
```

```python
np.save('embedings.npy', embedings)
files.download('embedings.npy')
```
